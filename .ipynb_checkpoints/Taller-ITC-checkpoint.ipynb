**3)** Sea $M$ un automata finito deterministico que acepta las palabras $u\in { \Sigma  }^{ * }$ definido como $M=(Q,\Sigma ,\delta ,{ q }_{ 0 },F)$, podemos construir un automa $M'$ definimo como $M'=(Q',\Sigma' ,\delta' ,{ q }_{ 0 }',F')$ tal que acepte las palabras $u\in { \Sigma  }^{ * }$ si y solo si $u$ no es aceptada por $M$.

Para emepezar es necesaria la creacion de un estado extra  $\{ { q }_{ e }\}$, luego definimos $M'$ de la siguiente manera.

$Q'=\{ Q\} \bigcup { \{  } { q }_{ e }\}$ 

$\Sigma '=\Sigma $

${ q }_{ 0 }'={ q }_{ 0 }$

$F'=q\in Q'\wedge q\notin F$

para los estados $q\in Q$ definimos la funcion de transición como:

$\delta'(q, a) =
\left\{
    \begin{array}{ll}
      { q }_{ e }\quad si\quad\delta(q, a) = \emptyset  \\
      \delta(q, a) \quad si\quad\delta(q, a)\neq \emptyset  \
    \end{array}
\right.$

para el unico esta faltante ${q}_{e}$ definimos la funcion de transición como:

$\delta'({ q }_{ e }, a) = { q }_{ e }$

de esta forma la funcion cubre todos los estados de $Q'$.


**4)** Empezamos definiendo del automa ${ M }_{ 1 }=({ Q }_{ 1 },\Sigma ,{ \delta  }_{ 1 },{ q }_{ 1 },{ F }_{ 1 })$ y el automa ${ M }_{ 2 }=({ Q }_{ 2 },\Sigma ,{ \delta  }_{ 2 },{ q }_{ 2 },{ F }_{ 2 })$, ahora para la construncion de un automata ${ M }=({ Q },{ \Sigma  }_{ m } ,{ \delta  },{ q }_{ 0 },{ F })$ usaremos la idea del automata producto, donde hacemos producto cruz entre ${ Q }_{ 1 }$ y ${ Q }_{ 2 }$, para definir los estados de $M$, de esta forma definimos el automa de la siguiente manera.

$Q={ Q }_{ 1 }\times Q_{ 2 }$

${ \Sigma  }_{ m } = \Sigma$

Para ${ q }_{ i }\in { Q }_{ 1 }$ y ${ q }_{ j }\in { Q }_{ 2 }$, definimos la funcion de transicion como:

$\delta (({ q }_{ i },{ q }_{ j }),a)=({ \delta  }_{ 1 }({ q }_{ i },a),{ \delta  }_{ 2 }({ q }_{ j },a))$

${ q }_{ 0 } = ({ q }_{ 1 }, { q }_{ 2 })$

$F = ({ q }_{ i },{ q }_{ j }) \quad t.q  \quad { q }_{ i }\in { F }_{ 1 } \wedge { q }_{ j }\in { F }_{ 2 }$

**8)** Empecemos notando que la palabra $ac\in L$, ahora el recorrido del automata para esta palabra se divide en dos hilos

a. 1 -> 2 -> 4

b. 1 -> 6 -> 5

Para el hilo a. se termino en el estado 4, para el hilo b. se termino en el etado 5, pero ni 4 ni 5 son estados de aceptación, por tanto el automa no acepta el lenguje $L$.
